```python
import os
import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

```python
import csv
```

```python
G = nx.DiGraph()

with open('20180321.csv', 'r') as input_file:
    lines = csv.reader(input_file)
    for line in lines:
        G.add_edge(int(line[0]), int(line[1]))
```

```python
pr=nx.pagerank(G,alpha=0.85)
```
