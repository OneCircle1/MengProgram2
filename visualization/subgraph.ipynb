```python
import networkx as nx
import csv
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
with open('20180321.csv', 'r') as input_file:
    lines = csv.reader(input_file)
    for line in lines:
        G.add_edge(int(line[0]), int(line[1]))
```

```python
nodelist = [n for n in G.nodes]
remove_list = []
```

```python
for i in range(len(nodelist)):
    if G.degree[nodelist[i]] == 1:
#     if G.degree[nodelist[i]] <= 10:
        remove_list.append(nodelist[i])
```

```python
G.remove_nodes_from(remove_list)
```

```python
subgraph = list(nx.connected_component_subgraphs(G))
```

```python
for i in range(len(subgraph)):
    if len(subgraph[i].nodes) >= 50:
        print(len(subgraph[i].nodes))
        nx.draw(subgraph[i], node_size = 1)
```
